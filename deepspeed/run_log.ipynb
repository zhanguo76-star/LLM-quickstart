# 3b
3b跑不了，主要是内存和gpu内存不足
正如刚才算的：T5-3B + CPU offload 需要 ~46GB RAM，你的机器只有 31GB，日志也证实了 CPU Virtual Memory: used = 29.29 GB, percent = 94.2% ——RAM 已经 94% 了然后被 OOM Killer 杀掉。

T5-3B 在 31GB RAM 的机器上确实跑不了 CPU offload。 换成 T5-large 是最合理的选择，
- T5-3B 如果要跑 ：至少需要 48GB RAM 的机器，或者 2 张 GPU 用 ZeRO-3 不做 CPU offload
